```python
import numpy as np
import os
np.random.seed(42)

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

def gaussian_rbf(x, landmark, gamma):
    return np.exp(-gamma * np.linalg.norm(x - landmark, axis=1)**2)
gamma = 10

x1s = np.linspace(-4.5, 4.5, 200).reshape(-1, 1)

plt.figure()
x2s = gaussian_rbf(x1s, 0, 0.1)
plt.plot(x1s, x2s)
x2s = gaussian_rbf(x1s, 0, 1)
plt.plot(x1s, x2s)
x2s = gaussian_rbf(x1s, 0, 10)
plt.plot(x1s, x2s)

plt.legend(['gamma = 0.1', 'gamma = 1', 'gamma = 10'])
plt.grid(True, which='both')
```

```python
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=400, factor=0.3, noise=0.1,random_state=42)

def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bo")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "go")
    plt.axis(axes)
    plt.gca().set_aspect('equal')
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

plot_dataset(X, y, [-1.5, 1.5, -1.5, 1.5])
plt.show()
```
